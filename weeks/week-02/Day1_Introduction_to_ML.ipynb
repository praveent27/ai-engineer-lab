## What is Machine Learning?

Machine Learning is learning patterns from data to make predictions or decisions without explicit programming.

Instead of writing rules like "if height > 180cm then basketball player", ML learns the relationship automatically from examples.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification, make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import accuracy_score, mean_squared_error

# Set random seed for reproducibility
np.random.seed(42)
plt.style.use('seaborn-v0_8')
sns.set_palette('husl')
```

## Supervised Learning

Training data has **labels/targets**. The algorithm learns to predict known outputs.

- **Regression**: Predict continuous values (house prices, temperature)
- **Classification**: Predict discrete categories (spam/not spam, medical diagnosis)

80% of real-world ML is supervised.

```python
# Generate synthetic regression data
X_reg, y_reg = make_regression(n_samples=100, n_features=1, noise=10, random_state=42)
X_reg_train, X_reg_test, y_reg_train, y_reg_test = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# Train linear regression
reg_model = LinearRegression()
reg_model.fit(X_reg_train, y_reg_train)

# Make predictions
y_reg_pred = reg_model.predict(X_reg_test)

print(f"Regression MSE: {mean_squared_error(y_reg_test, y_reg_pred):.2f}")

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(X_reg_test, y_reg_test, alpha=0.7, label='True values')
plt.plot(X_reg_test, y_reg_pred, color='red', linewidth=2, label='Predictions')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Supervised Learning: Regression Example')
plt.legend()
plt.show()
```

```python
# Generate synthetic classification data
X_clf, y_clf = make_classification(n_samples=100, n_features=2, n_classes=2, 
                                  n_redundant=0, n_clusters_per_class=1, random_state=42)
X_clf_train, X_clf_test, y_clf_train, y_clf_test = train_test_split(X_clf, y_clf, test_size=0.2, random_state=42)

# Train logistic regression
clf_model = LogisticRegression()
clf_model.fit(X_clf_train, y_clf_train)

# Make predictions
y_clf_pred = clf_model.predict(X_clf_test)

print(f"Classification Accuracy: {accuracy_score(y_clf_test, y_clf_pred):.2f}")

# Plot decision boundary
plt.figure(figsize=(10, 6))
plt.scatter(X_clf_test[:, 0], X_clf_test[:, 1], c=y_clf_test, cmap='coolwarm', alpha=0.7)

# Create meshgrid for decision boundary
x_min, x_max = X_clf_test[:, 0].min() - 1, X_clf_test[:, 0].max() + 1
y_min, y_max = X_clf_test[:, 1].min() - 1, X_clf_test[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
Z = clf_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Supervised Learning: Classification Example')
plt.show()
```

## Unsupervised Learning

No labels — the algorithm finds **hidden structure** in the data.

- **Clustering**: Group similar items (customer segmentation)
- **Dimensionality Reduction**: Compress data while preserving structure (PCA)

Used for exploration and preprocessing.

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Generate synthetic clustering data
X_unsup, _ = make_classification(n_samples=150, n_features=4, n_classes=3, 
                                n_redundant=0, n_clusters_per_class=1, random_state=42)

# K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_unsup)

# PCA for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_unsup)

plt.figure(figsize=(12, 5))

# Original data (PCA projection)
plt.subplot(1, 2, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.7)
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('Original Data (PCA Projection)')

# Clustered data
plt.subplot(1, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', alpha=0.7)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
           c='red', marker='x', s=200, linewidth=3, label='Centroids')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('K-Means Clustering')
plt.legend()

plt.tight_layout()
plt.show()

print(f"Explained variance by PCA components: {pca.explained_variance_ratio_}")
```

## Key Takeaways

- **Supervised Learning**: Labeled data → prediction tasks (regression/classification)
- **Unsupervised Learning**: Unlabeled data → pattern discovery (clustering/dimensionality reduction)
- ML learns from data, not code — the algorithm finds patterns humans might miss
- Most business problems are supervised: predict sales, detect fraud, recommend products
- Choose algorithm based on data type and problem type, not the other way around
