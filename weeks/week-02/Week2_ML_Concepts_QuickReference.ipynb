## Linear Algebra Cheat Sheet

### Matrix Operations
- **Matrix Multiplication**: C = A·B (m×n · n×p = m×p)
- **Transpose**: (A·B)ᵀ = Bᵀ·Aᵀ
- **Inverse**: A⁻¹·A = I (only square matrices)
- **Determinant**: det(A) measures volume scaling

### Vector Operations
- **Dot Product**: a·b = Σaᵢbᵢ = |a||b|cosθ
- **Norm**: ||v||₂ = √(Σvᵢ²)
- **Unit Vector**: û = v/||v||

### Eigenvalues & Eigenvectors
- **Definition**: A·v = λ·v
- **Characteristic Equation**: det(A - λI) = 0
- **Spectral Theorem**: Symmetric matrices diagonalizable

### Key Concepts
- **Rank**: Maximum linearly independent rows/columns
- **Null Space**: Solutions to A·x = 0
- **Column Space**: Span of matrix columns
- **Singular Values**: Square roots of eigenvalues of AᵀA

## Statistics for ML Cheat Sheet

### Probability Distributions
- **Normal/Gaussian**: N(μ, σ²) — Central Limit Theorem
- **Bernoulli**: P(X=1)=p — Binary outcomes
- **Binomial**: B(n,p) — Sum of n Bernoulli trials
- **Poisson**: P(X=k)=e^{-λ}λᵏ/k! — Count events

### Statistical Tests
- **t-test**: Compare means of two groups
- **Chi-square**: Test independence of categorical variables
- **ANOVA**: Compare means of 3+ groups
- **Correlation**: Pearson (linear), Spearman (rank)

### Key Formulas
- **Mean**: μ = Σxᵢ/n
- **Variance**: σ² = Σ(xᵢ-μ)²/n
- **Standard Deviation**: σ = √σ²
- **Covariance**: Cov(X,Y) = Σ(xᵢ-μx)(yᵢ-μy)/n
- **Correlation**: ρ = Cov(X,Y)/(σx·σy)

### Hypothesis Testing
- **Null Hypothesis (H₀)**: No effect/difference
- **Alternative Hypothesis (H₁)**: Effect exists
- **p-value**: Probability of data given H₀ true
- **Significance Level (α)**: Usually 0.05
- **Power**: 1 - P(Type II error)

## Model Evaluation Cheat Sheet

### Classification Metrics
- **Accuracy**: (TP+TN)/(TP+TN+FP+FN)
- **Precision**: TP/(TP+FP) — Quality
- **Recall**: TP/(TP+FN) — Completeness
- **F1**: 2·P·R/(P+R) — Harmonic mean
- **Specificity**: TN/(TN+FP) — True negative rate

### Regression Metrics
- **MAE**: Σ|y-ŷ|/n — Mean absolute error
- **MSE**: Σ(y-ŷ)²/n — Mean squared error
- **RMSE**: √MSE — Root mean squared error
- **R²**: 1 - SS_res/SS_tot — Explained variance

### ROC & AUC
- **ROC Curve**: TPR vs FPR at different thresholds
- **AUC**: Area under ROC curve (0.5=random, 1.0=perfect)
- **Interpretation**: AUC > 0.8 good, > 0.9 excellent

### Cross-Validation
- **K-Fold CV**: Split into K folds, train on K-1, test on 1
- **Stratified CV**: Maintain class proportions
- **Leave-One-Out**: K=n (expensive but unbiased)

````python
# Quick evaluation example
y_true = np.array([0, 0, 1, 1, 1, 1])
y_pred = np.array([0, 1, 1, 1, 0, 1])

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:")
print(cm)

# Manual calculations
TN, FP, FN, TP = cm.ravel()
accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1 = 2 * precision * recall / (precision + recall)

print(f"\nAccuracy: {accuracy:.3f}")
print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
print(f"F1 Score: {f1:.3f}")

# Plot confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.title('Confusion Matrix Example')
plt.show()
````

## Linear Regression Cheat Sheet

### Model
- **Equation**: ŷ = w·x + b
- **Cost Function**: J(w,b) = (1/2n)Σ(y - ŷ)²
- **Gradients**: ∂J/∂w = (1/n)Xᵀ·(ŷ - y), ∂J/∂b = (1/n)Σ(ŷ - y)

### Normal Equation
- **Closed Form**: w = (XᵀX)⁻¹Xᵀy
- **Pros**: No iteration, exact solution
- **Cons**: O(n³) complexity, doesn't scale

### Gradient Descent
- **Update**: w = w - α·∂J/∂w
- **Learning Rate**: α (0.001 to 0.1 typically)
- **Convergence**: Cost stops decreasing
- **Feature Scaling**: Essential for convergence

### Regularisation
- **L2 (Ridge)**: J = MSE + λ·Σw²
- **L1 (Lasso)**: J = MSE + λ·Σ|w|
- **Elastic Net**: Combination of L1 + L2

### Assumptions
- Linear relationship between x and y
- Independent observations
- Homoscedasticity (constant variance)
- No multicollinearity

## Logistic Regression Cheat Sheet

### Model
- **Equation**: P(y=1|x) = σ(w·x + b) = 1/(1+e^(-(w·x+b)))
- **Decision Rule**: ŷ = 1 if P(y=1|x) ≥ 0.5 else 0

### Cost Function
- **Binary Cross-Entropy**: L = -Σ[y·log(ŷ) + (1-y)·log(1-ŷ)]
- **Why not MSE?**: Non-convex, slow convergence

### Gradients
- **∂L/∂w**: (1/n)Xᵀ·(ŷ - y)
- **∂L/∂b**: (1/n)Σ(ŷ - y)
- **Same as linear regression!**

### Multiclass (Softmax)
- **Softmax**: P(y=k|x) = e^(z_k)/Σe^(z_j)
- **Cross-Entropy**: L = -ΣΣy_k·log(P(y=k|x))

### Key Properties
- Outputs probabilities (not raw scores)
- Linear decision boundaries
- Interpretable coefficients
- Robust to noise

## Cross-Validation & Tuning Cheat Sheet

### CV Methods
- **K-Fold**: Split into K folds, K-1 train, 1 test
- **Stratified K-Fold**: Maintain class proportions
- **Leave-One-Out**: K=n, expensive but unbiased
- **Time Series**: Rolling/expanding windows

### Hyperparameter Tuning
- **Grid Search**: Exhaustive search over parameter grid
- **Random Search**: Sample random combinations
- **Bayesian Optimization**: Use previous results to guide search

### Validation Curves
- **Bias-Variance**: Plot train/val score vs hyperparameter
- **Underfitting**: Both scores low
- **Overfitting**: Train high, validation low
- **Optimal**: Validation score peaks

### Learning Curves
- **Plot**: Score vs training set size
- **Gap Analysis**: Large gap = overfitting
- **Plateau**: More data won't help
- **Slope**: Still learning from data

## ML Pipeline Cheat Sheet

### 1. Data Collection
- Load data (CSV, database, API)
- Check data types and basic stats
- Identify target variable

### 2. Data Cleaning
- Handle missing values (drop, impute, flag)
- Remove duplicates
- Detect and handle outliers
- Fix data types

### 3. Feature Engineering
- Create new features from existing
- Encode categorical variables
- Scale/normalize features
- Handle multicollinearity

### 4. Model Training
- Split data (train/val/test)
- Choose appropriate models
- Train with cross-validation
- Monitor for overfitting

### 5. Model Evaluation
- Use multiple metrics (not just accuracy)
- Plot confusion matrix, ROC curves
- Check for bias/fairness issues
- Validate on unseen data

### 6. Hyperparameter Tuning
- Use grid/random search with CV
- Focus on most important parameters
- Validate improvements

### 7. Model Selection
- Compare models on validation metrics
- Consider computational cost
- Choose based on business requirements

### 8. Deployment
- Save model and preprocessing
- Create prediction API
- Monitor performance in production
- Plan for model updates

## Common ML Mistakes & Solutions

### Data Leakage
- **Problem**: Test information leaks into training
- **Solution**: Preprocessing after train/test split
- **Example**: Scaling with test set statistics

### Overfitting
- **Problem**: Model memorizes training data
- **Solution**: Cross-validation, regularisation, early stopping
- **Detection**: Train score >> validation score

### Underfitting
- **Problem**: Model too simple for data
- **Solution**: More complex model, better features
- **Detection**: Both train and validation scores low

### Imbalanced Data
- **Problem**: Rare events poorly predicted
- **Solution**: Stratified CV, class weights, resampling
- **Metrics**: Use F1, AUC instead of accuracy

### Non-stationary Data
- **Problem**: Data distribution changes over time
- **Solution**: Time-based splits, concept drift detection
- **Monitoring**: Track performance metrics over time
