## 1. Matplotlib Fundamentals

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Set a clean style
plt.style.use('seaborn-v0_8-whitegrid')

# If in Jupyter, this makes plots appear inline:
# %matplotlib inline
```

```python
# ── Basic Line Plot ──
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, y1, label='sin(x)', color='#2E75B6', linewidth=2)
ax.plot(x, y2, label='cos(x)', color='#ED7D31', linewidth=2, linestyle='--')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_title('Sine and Cosine Functions', fontsize=14, fontweight='bold')
ax.legend(fontsize=11)
ax.axhline(y=0, color='grey', linewidth=0.5)
plt.tight_layout()
plt.savefig('line_plot.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: line_plot.png")
```

```python
# ── Histogram — Distribution of Data ──
np.random.seed(42)
data = np.random.randn(1000)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Standard histogram
axes[0].hist(data, bins=30, color='#2E75B6', edgecolor='white', alpha=0.8)
axes[0].axvline(data.mean(), color='red', linestyle='--', label=f'Mean: {data.mean():.2f}')
axes[0].set_title('Normal Distribution', fontweight='bold')
axes[0].legend()

# KDE overlay
axes[1].hist(data, bins=30, density=True, color='#2E75B6', edgecolor='white', alpha=0.6)
x_kde = np.linspace(-4, 4, 100)
axes[1].plot(x_kde, (1/np.sqrt(2*np.pi)) * np.exp(-x_kde**2/2), 'r-', linewidth=2, label='Normal PDF')
axes[1].set_title('With PDF Overlay', fontweight='bold')
axes[1].legend()

plt.tight_layout()
plt.savefig('histogram.png', dpi=150)
plt.show()
```

```python
# ── Scatter Plot — Relationship Between Variables ──
np.random.seed(42)
n = 200
x = np.random.randn(n)
y = 0.5 * x + np.random.randn(n) * 0.5
colors = np.random.randn(n)

fig, ax = plt.subplots(figsize=(8, 6))
scatter = ax.scatter(x, y, c=colors, cmap='RdYlBu', alpha=0.7, edgecolors='white', s=60)
plt.colorbar(scatter, label='Category Score')
ax.set_xlabel('Feature X')
ax.set_ylabel('Feature Y')
ax.set_title('Scatter Plot with Colour Mapping', fontweight='bold')

# Add trend line
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
x_line = np.linspace(x.min(), x.max(), 100)
ax.plot(x_line, p(x_line), 'r--', linewidth=2, label=f'Trend: y={z[0]:.2f}x+{z[1]:.2f}')
ax.legend()
plt.tight_layout()
plt.savefig('scatter.png', dpi=150)
plt.show()
```

```python
# ── Subplots Grid — Multiple Charts ──
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 32]
axes[0, 0].bar(categories, values, color=['#2E75B6', '#ED7D31', '#548235', '#7030A0', '#BF0000'])
axes[0, 0].set_title('Bar Chart', fontweight='bold')

# Pie chart
axes[0, 1].pie(values, labels=categories, autopct='%1.0f%%', colors=['#2E75B6', '#ED7D31', '#548235', '#7030A0', '#BF0000'])
axes[0, 1].set_title('Pie Chart', fontweight='bold')

# Box plot
data = [np.random.randn(100) + i for i in range(5)]
bp = axes[1, 0].boxplot(data, labels=categories, patch_artist=True)
for patch, color in zip(bp['boxes'], ['#2E75B6', '#ED7D31', '#548235', '#7030A0', '#BF0000']):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)
axes[1, 0].set_title('Box Plot', fontweight='bold')

# Heatmap (manual)
matrix = np.random.rand(5, 5)
im = axes[1, 1].imshow(matrix, cmap='YlOrRd')
axes[1, 1].set_title('Heatmap', fontweight='bold')
plt.colorbar(im, ax=axes[1, 1])

plt.suptitle('Chart Types Gallery', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('chart_gallery.png', dpi=150)
plt.show()
```

## 2. Seaborn — Statistical Visualisation

Seaborn wraps Matplotlib with beautiful defaults and statistical plot types.

```python
import seaborn as sns

# Load built-in dataset (Titanic — we'll use this tomorrow!)
# For now, create synthetic data
np.random.seed(42)
n = 300
df = pd.DataFrame({
    'age': np.random.normal(35, 12, n).clip(18, 80).astype(int),
    'salary': np.random.normal(70000, 25000, n).clip(25000, 200000).astype(int),
    'experience': np.random.randint(0, 30, n),
    'department': np.random.choice(['Eng', 'Mkt', 'Sales', 'HR', 'Finance'], n),
    'performance': np.random.choice(['Low', 'Medium', 'High'], n, p=[0.2, 0.5, 0.3])
})
df['salary'] = df['salary'] + df['experience'] * 2000 + np.random.randn(n) * 5000

print(df.head())
print(f"\nShape: {df.shape}")
```

```python
# ── Distribution Plots ──
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Histogram + KDE
sns.histplot(df['salary'], kde=True, ax=axes[0], color='#2E75B6')
axes[0].set_title('Salary Distribution')

# Box plot by category
sns.boxplot(x='department', y='salary', data=df, ax=axes[1], palette='Set2')
axes[1].set_title('Salary by Department')
axes[1].tick_params(axis='x', rotation=45)

# Violin plot (box + KDE)
sns.violinplot(x='performance', y='salary', data=df, ax=axes[2], palette='muted',
               order=['Low', 'Medium', 'High'])
axes[2].set_title('Salary by Performance')

plt.tight_layout()
plt.savefig('seaborn_distributions.png', dpi=150)
plt.show()
```

```python
# ── Correlation Heatmap (ESSENTIAL for feature selection) ──
numeric_cols = df[['age', 'salary', 'experience']]
corr = numeric_cols.corr()

fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='RdBu_r', center=0, fmt='.2f',
            square=True, linewidths=1, ax=ax)
ax.set_title('Correlation Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('correlation_heatmap.png', dpi=150)
plt.show()

# 💡 Highly correlated features (>0.8 or <-0.8) should be investigated
# — you may need to drop one to avoid multicollinearity
```

```python
# ── Pair Plot — All Relationships at Once ──
# Great for EDA: see distributions + relationships in one view
g = sns.pairplot(df[['age', 'salary', 'experience', 'performance']], 
                 hue='performance', palette='Set1', diag_kind='kde',
                 plot_kws={'alpha': 0.5, 's': 30})
g.figure.suptitle('Pair Plot: Numeric Features by Performance', y=1.02, fontweight='bold')
plt.savefig('pairplot.png', dpi=100)
plt.show()
```

```python
# ── Count Plot & Cat Plot ──
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Count plot (like a bar chart for categories)
sns.countplot(x='department', hue='performance', data=df, ax=axes[0], palette='Set2',
              order=df['department'].value_counts().index)
axes[0].set_title('Headcount by Dept & Performance')
axes[0].tick_params(axis='x', rotation=45)

# Bar plot with confidence intervals
sns.barplot(x='department', y='salary', data=df, ax=axes[1], palette='Blues_d',
            order=['Eng', 'Finance', 'Mkt', 'Sales', 'HR'], errorbar='sd')
axes[1].set_title('Mean Salary by Department')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('categorical_plots.png', dpi=150)
plt.show()
```

## 3. ML-Specific Visualisations

These chart types come up constantly when building and evaluating models.

```python
# ── Simulated ML metrics for visualisation ──

# Loss curve (training progress)
epochs = np.arange(1, 51)
train_loss = 2.5 * np.exp(-0.08 * epochs) + 0.3 + np.random.randn(50) * 0.05
val_loss = 2.5 * np.exp(-0.06 * epochs) + 0.45 + np.random.randn(50) * 0.08

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Training curves
axes[0].plot(epochs, train_loss, label='Train Loss', color='#2E75B6', linewidth=2)
axes[0].plot(epochs, val_loss, label='Validation Loss', color='#ED7D31', linewidth=2)
axes[0].fill_between(epochs, train_loss, val_loss, alpha=0.1, color='red')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].set_title('Training vs Validation Loss', fontweight='bold')
axes[0].legend()
axes[0].annotate('Overfitting starts here', xy=(25, 0.6), xytext=(30, 1.0),
                arrowprops=dict(arrowstyle='->', color='red'), color='red', fontsize=10)

# Confusion matrix
from matplotlib.colors import LinearSegmentedColormap
confusion = np.array([[85, 10, 5],
                      [8, 78, 14],
                      [3, 12, 85]])
labels = ['Cat', 'Dog', 'Bird']
sns.heatmap(confusion, annot=True, fmt='d', cmap='Blues', 
            xticklabels=labels, yticklabels=labels, ax=axes[1])
axes[1].set_xlabel('Predicted')
axes[1].set_ylabel('Actual')
axes[1].set_title('Confusion Matrix', fontweight='bold')

plt.tight_layout()
plt.savefig('ml_visualisations.png', dpi=150)
plt.show()
```

```python
# ── Feature Importance Plot ──
features = ['experience', 'age', 'dept_encoded', 'education', 'certifications', 
            'location', 'gender', 'team_size']
importance = np.array([0.32, 0.25, 0.15, 0.10, 0.08, 0.05, 0.03, 0.02])

# Sort by importance
idx = np.argsort(importance)

fig, ax = plt.subplots(figsize=(8, 6))
ax.barh([features[i] for i in idx], importance[idx], color='#2E75B6')
ax.set_xlabel('Importance Score')
ax.set_title('Feature Importance (Random Forest)', fontweight='bold')
for i, v in enumerate(importance[idx]):
    ax.text(v + 0.005, i, f'{v:.2f}', va='center', fontsize=10)
plt.tight_layout()
plt.savefig('feature_importance.png', dpi=150)
plt.show()
```

## 4. Quick Reference

| Plot Type | When to Use | Seaborn Function |
|-----------|-------------|------------------|
| Histogram | Distribution of one variable | `sns.histplot()` |
| Box plot | Distribution + outliers by group | `sns.boxplot()` |
| Violin | Distribution shape by group | `sns.violinplot()` |
| Scatter | Relationship between 2 variables | `sns.scatterplot()` |
| Heatmap | Correlation matrix, confusion matrix | `sns.heatmap()` |
| Pair plot | All pairwise relationships | `sns.pairplot()` |
| Bar / Count | Categorical comparisons | `sns.barplot()` / `sns.countplot()` |
| Line | Time series, training curves | `plt.plot()` |

**Styling tips:**
- Always add titles, axis labels, and legends
- Use `plt.tight_layout()` to prevent overlap
- Save with `dpi=150` for crisp images
- Use `figsize=(width, height)` to control size
