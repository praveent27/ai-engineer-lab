## NumPy Cheat Sheet

```python
import numpy as np

# ═══ CREATION ═══
np.array([1,2,3])                  # from list
np.zeros((3,4))                    # 3x4 zeros
np.ones((2,3))                     # 2x3 ones
np.full((2,2), 7)                  # 2x2 filled with 7
np.eye(3)                          # 3x3 identity
np.arange(0, 10, 2)               # [0,2,4,6,8]
np.linspace(0, 1, 5)              # 5 even points 0→1
np.random.seed(42)                 # reproducibility
np.random.rand(3,3)               # uniform [0,1)
np.random.randn(3,3)              # normal (0,1)
np.random.randint(1, 100, (2,5))  # random ints

# ═══ PROPERTIES ═══
a = np.array([[1,2],[3,4]])
a.shape                            # (2,2)
a.dtype                            # int64
a.size                             # 4
a.ndim                             # 2

# ═══ INDEXING ═══
a[0, 1]                            # element at row 0, col 1
a[0]                               # row 0
a[:, 1]                            # col 1
a[0:2, 1:3]                       # submatrix
a[a > 2]                          # boolean filtering
np.where(a > 2, 'high', 'low')   # conditional

# ═══ OPERATIONS ═══
a + b; a - b; a * b; a / b       # element-wise
a @ b                              # matrix multiply
a.T                                # transpose
a.flatten()                        # to 1D
a.reshape(1, -1)                  # auto-calc shape
np.dot(a, b)                      # dot product

# ═══ AGGREGATION ═══
a.sum(); a.mean(); a.std(); a.var()
a.min(); a.max(); a.argmin(); a.argmax()
np.median(a); np.percentile(a, 75)
a.sum(axis=0)                     # column sums
a.sum(axis=1)                     # row sums

# ═══ LINEAR ALGEBRA ═══
np.linalg.norm(v)                 # vector magnitude
np.linalg.inv(A)                  # matrix inverse
np.linalg.eig(A)                  # eigenvalues/vectors
np.linalg.det(A)                  # determinant

# ═══ STACKING ═══
np.vstack([a, b])                 # vertical
np.hstack([a, b])                 # horizontal
np.concatenate([a, b], axis=0)    # along axis
```
