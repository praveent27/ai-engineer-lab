# Day 2 — Data Preprocessing Review
This notebook revisits core preprocessing techniques from Week 3 and applies them in a capstone-oriented pipeline.

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
numeric_features = ['feature_a', 'feature_b']
categorical_features = ['category_a', 'category_b']

numeric_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler()),
])
categorical_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False)),
])

preprocessor = ColumnTransformer([
    ('num', numeric_transformer, numeric_features),
    ('cat', categorical_transformer, categorical_features),
])
preprocessor
```

## Notes
- Impute numeric values with median if outliers are present.
- Use sparse_output=False for one-hot encoding when you need a dense matrix.
- Keep categorical encoders inside the pipeline to avoid data leakage.
