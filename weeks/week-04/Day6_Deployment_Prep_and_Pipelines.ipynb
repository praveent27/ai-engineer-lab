# Day 6 — Deployment Prep and Pipelines
This notebook prepares the trained pipeline for production, including serialization and an inference wrapper.

````python
import pandas as pd
import joblib
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
````

````python
pipeline = Pipeline([
    ('model', LinearRegression()),
])

joblib.dump(pipeline, 'week4_pipeline.pkl')
````

## Inference wrapper
```python
def predict(raw_df):
    pipeline = joblib.load('week4_pipeline.pkl')
    return pipeline.predict(raw_df)
```
Keep the wrapper small and validation-aware.
